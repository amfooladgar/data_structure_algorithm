# 767 Reorganize String
Given a string s, rearrange the characters of s so that any two adjacent characters are not the same.

Return any possible rearrangement of s or return "" if not possible.

 

Example 1:
```
Input: s = "aab"
Output: "aba"
```
Example 2:
```
Input: s = "aaab"
Output: ""
```

Constraints:

- 1 <= s.length <= 500
- s consists of lowercase English letters.

Hint:
- Alternate placing the most common letters.

````python
import heapq

def reorganizeString(s):
    # Step 1: Count frequency of each character
    freq_map = [0] * 26
    for ch in s:
        freq_map[ord(ch) - ord('a')] += 1

    max_count = max(freq_map)
    n = len(s)
    
    # Step 2: Early check â€“ if the most frequent char is too frequent, return ""
    if max_count > (n + 1) // 2:
        return ""

    # Step 3: Create a max heap of (-freq, char)
    max_heap = [(-freq_map[i], chr(i + ord('a'))) for i in range(26) if freq_map[i] > 0]
    heapq.heapify(max_heap)

    res = []
    prev_count, prev_char = 0, ''

    # Step 4: Build the result string
    while max_heap:
        count, char = heapq.heappop(max_heap)
        res.append(char)

        # If there's a previous character held back, push it back into the heap
        if prev_count < 0:
            heapq.heappush(max_heap, (prev_count, prev_char))

        # Hold the current char out for one round (decrement its count)
        prev_count, prev_char = count + 1, char

    return ''.join(res)
````

````python
s = "aaabccs"
reorganizeString(s)
````
