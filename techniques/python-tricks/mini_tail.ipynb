# Mini tail -f
Implement a CLI that prints the last N lines of a file and optionally follows the file as new lines are appended.

Example usage:

python tail.py -n 20 app.log
python tail.py -f app.log

Goal:

- Print the last N lines of a file.
- Optionally follow the file as new lines are appended.
- Avoid loading huge files into memory.
- Handle Ctrl+C.
- Bonus: handle log rotation/truncation.
