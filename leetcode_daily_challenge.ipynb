# 2348. Number of Zero-Filled Subarrays - Aug19

```python
def zeroFilledSubarray(nums):
    count_prev, total_subarrays = 0, 0
    n = len(nums)
    i = 0
    while i<n:
        if nums[i]==0:
            count_prev+=1
            total_subarrays+=count_prev

        else:
            count_prev = 0
        i+=1
    
    return total_subarrays         
```

```python
nums = [1,3,0,0,2,0,0,4]
zeroFilledSubarray(nums)
```

```python
nums = [0,0,0,2,0,0]
zeroFilledSubarray(nums)
```

```python
# Another solution 
def zeroFilledSubarray(nums):
    size, ans = 0, 0
    for num in nums:
        size = size + 1 if not num else 0
        ans += size
    return ans
```

```python
nums = [0,0,0,2,0,0]
zeroFilledSubarray(nums)
```

```python
# Another solution
def zeroFilledSubarray(nums):
    start = subarrays = 0
    for end in range(len(nums)):
        if nums[end]:
            start = end + 1
        subarrays += end + 1 - start
    return subarrays
```

```python
nums = [1,3,0,0,2,0,0,4]
zeroFilledSubarray(nums)
```
