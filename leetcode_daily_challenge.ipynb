# 2348. Number of Zero-Filled Subarrays - Aug19-2025

````python
def zeroFilledSubarray(nums):
    count_prev, total_subarrays = 0, 0
    n = len(nums)
    i = 0
    while i<n:
        if nums[i]==0:
            count_prev+=1
            total_subarrays+=count_prev

        else:
            count_prev = 0
        i+=1
    
    return total_subarrays         
````

````python
nums = [1,3,0,0,2,0,0,4]
zeroFilledSubarray(nums)
````

````python
nums = [0,0,0,2,0,0]
zeroFilledSubarray(nums)
````

````python
# Another solution 
def zeroFilledSubarray(nums):
    size, ans = 0, 0
    for num in nums:
        size = size + 1 if not num else 0
        ans += size
    return ans
````

````python
nums = [0,0,0,2,0,0]
zeroFilledSubarray(nums)
````

````python
# Another solution
def zeroFilledSubarray(nums):
    start = subarrays = 0
    for end in range(len(nums)):
        if nums[end]:
            start = end + 1
        subarrays += end + 1 - start
    return subarrays
````

# 679. 24 Game, Aug18,2025

You are given an integer array cards of length 4. You have four cards, each containing a number in the range [1, 9]. You should arrange the numbers on these cards in a mathematical expression using the operators ['+', '-', '*', '/'] and the parentheses '(' and ')' to get the value 24.

You are restricted with the following rules:

The division operator '/' represents real division, not integer division.
For example, 4 / (1 - 2 / 3) = 4 / (1 / 3) = 12.
Every operation done is between two numbers. In particular, we cannot use '-' as a unary operator.
For example, if cards = [1, 1, 1, 1], the expression "-1 - 1 - 1 - 1" is not allowed.
You cannot concatenate numbers together
For example, if cards = [1, 2, 1, 2], the expression "12 + 12" is not valid.
Return true if you can get such expression that evaluates to 24, and false otherwise.

 

Example 1:

Input: cards = [4,1,8,7]
Output: true
Explanation: (8-4) * (7-1) = 24
Example 2:

Input: cards = [1,2,1,2]
Output: false
 

Constraints:

cards.length == 4
1 <= cards[i] <= 9

````python
# Using backtracking
def judgePoint24(cards):
    EPS = 1e-6
    nums = [float(x) for x in cards]

    def generate_possible_results(x, y):
        res = {x + y, x - y, y - x, x * y}
        if abs(x) > EPS:  # denominator for y/x is x
            res.add(y / x)
        if abs(y) > EPS:  # denominator for x/y is y
            res.add(x / y)
        return res

    def backtrack(arr):
        if len(arr) == 1:
            return abs(arr[0] - 24.0) <= EPS

        n = len(arr)
        for i in range(n):
            for j in range(i + 1, n):
                # pick a pair once (i < j)
                a, b = arr[i], arr[j]
                rest = [arr[k] for k in range(n) if k != i and k != j]
                for r in generate_possible_results(a, b):
                    if backtrack(rest + [r]):
                        return True
        return False

    return backtrack(nums)
````

````python
cards = [4,1,8,7]
judgePoint24(cards)
````

````python
# Slightly faster with memoization
def judgePoint24(cards):
    EPS = 1e-6
    nums = [float(x) for x in cards]
    seen = set()

    def key(arr):
        # round to stabilize floats; sort to make a canonical multiset
        return tuple(sorted(round(x, 4) for x in arr))

    def generate_possible_results(x, y):
        res = {x + y, x - y, y - x, x * y}
        if abs(x) > EPS:
            res.add(y / x)
        if abs(y) > EPS:
            res.add(x / y)
        return res

    def backtrack(arr):
        if len(arr) == 1:
            return abs(arr[0] - 24.0) <= EPS

        k = key(arr)
        if k in seen:
            return False
        seen.add(k)

        n = len(arr)
        for i in range(n):
            for j in range(i + 1, n):
                a, b = arr[i], arr[j]
                rest = [arr[k] for k in range(n) if k != i and k != j]
                for r in generate_possible_results(a, b):
                    if backtrack(rest + [r]):
                        return True
        return False

    return backtrack(nums)
````

````python
# using recursive approach
from functools import lru_cache
def countSquares(matrix):

    m = len(matrix)
    n = len(matrix[0])
    total = 0
    @lru_cache
    def dp(i,j):
        nonlocal total
        if i ==m or j ==n:
            return 0
        if matrix[i][j] == 1:
            size = 1 + min(dp(i+1,j), dp(i,j+1), dp(i+1,j+1))
            total+=size
            return size
        return 0
    
    for i in range(m):
        for j in range(n):
            dp(i,j)
    return total
        

````

````python
matrix =[
  [0,1,1,1],
  [1,1,1,1],
  [0,1,1,1]
]
countSquares(matrix)
````

````python
# using iterative bottom-up approach
def countSquares(matrix):

    m = len(matrix)
    n = len(matrix[0])
    dp = [[0]*(n+1) for _ in range(m+1)]
    total = 0
    for i in range(1,m+1):
        for j in range(1,n+1):
            if matrix[i-1][j-1] == 1:
                dp[i][j] = 1+ min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1])
                total+= dp[i][j]
            else:
                dp[i][j] = 0
    return total
                
            
````
