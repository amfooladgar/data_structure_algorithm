```python
class Solution(object):
    def firstUniqChar(self, s):
        temp = []
        j=0
        k=-1
        for i in range(len(s)):
            if  ((s[i] in s[i+1:]) or (s[i] in temp)):
                temp.append(s[i])
                continue
            else:
                return i
        return -1
```

```python
class Solution(object):
    def firstUniqChar(self, s):
        count = collections.Counter(s)
        print(count)
        print([(i,j) for i,j in enumerate(s)])
        # find the index
        for idx, ch in enumerate(s):
            if count[ch] == 1:
                return idx     
        return -1
```
