```python
def find_hidden_visible_dig(n):
    # result variable is to collect all the hidden and visible digits
    result = []
    # hidden_let to keep the mapping between the hidden lower-case letters 'a'-'j' with proper numbers
    hidden_let = {'a':'0','b':'1','c':'2','d':'3','e':'4','f':'5','g':'6','h':'7','i':'8','j':'9'}
    # a loop to go through the string
    for i in n:
        if(i.isdigit()):
            result.append(i)
        elif(i in hidden_let.keys()):
            result.append(hidden_let.get(i))
    if (len(result)==0):
        result = None
    return result
```
